```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/drive/MyDrive/housing.csv")
data.head()
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
features = data.drop(['median_house_value'],axis=1)
labels = data['median_house_value']
print(features,labels)
```

```python
features.hist(figsize=(15,8))
plt.show()
```

```python
features['total_rooms'] = np.log(features['total_rooms']+1)
features['total_bedrooms'] = np.log(features['total_bedrooms']+1)
features['population'] = np.log(features['population']+1)
features['households'] = np.log(features['households']+1)
features['median_income'] = np.log(features['median_income']+1)

features.hist(figsize=(15,8))
plt.show()
```

```python
data_scaled = data.copy(deep=True)
data_scaled['total_rooms'] = np.log(data_scaled['total_rooms']+1)
data_scaled['total_bedrooms'] = np.log(data_scaled['total_bedrooms']+1)
data_scaled['population'] = np.log(data_scaled['population']+1)
data_scaled['households'] = np.log(data_scaled['households']+1)
data_scaled['median_income'] = np.log(data_scaled['median_income']+1)

data_scaled.head()
fig, ax = plt.subplots(figsize=(10,5))
sns.heatmap(data_scaled.corr(),annot=True,cmap="YlGnBu",ax = ax)
plt.show()
```

```python
data_scaled.head()
```

```python
data_scaled['ocean_proximity'].value_counts()
```

```python
data_scaled = pd.concat([data_scaled,pd.get_dummies(data_scaled['ocean_proximity'])],axis=1)
data_scaled.drop('ocean_proximity',inplace=True,axis=1)
data_scaled.head()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.heatmap(data_scaled.corr(),annot=True,cmap="YlGnBu",ax = ax)
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='latitude',y='longitude',data=data_scaled,hue="median_house_value",palette="coolwarm")
plt.show()
```
