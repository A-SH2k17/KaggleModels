```python
df = pd.read_csv('/kaggle/input/housesalesprediction/kc_house_data.csv')
df
```

```python
df.info()
```

```python
features= ['bedrooms','bathrooms','sqft_living',
           'sqft_lot','floors','waterfront','view'
           ,'condition','grade','sqft_above','sqft_basement',
           'yr_built','yr_renovated','lat','long'
           ]
x = df[features]
y =df['price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split as tts
xtrain,xtest,ytrain,ytest = tts(x,y,test_size=0.3)
```

```python
xtrain
```

# ****Using simple linear regression**** #

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(xtrain,ytrain)
lr_accuracy=lr_model.score(xtest,ytest)
print(f'The accuracy of the simple linear regression model is {lr_accuracy}' )
```

# **Using Ridge Regression**

```python
from sklearn.linear_model import Ridge
r_model = Ridge()
r_model.fit(xtrain,ytrain)
r_accuracy=r_model.score(xtest,ytest)
print(f'The accuracy of the ridge regression model is {r_accuracy}' )
```

# **Using Random Forest Regressor**

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor()
rf_model.fit(xtrain,ytrain)
rf_accuracy=rf_model.score(xtest,ytest)
print(f'The accuracy of the random forest regression model is {rf_accuracy}' )
```

# **Using Elastic Net**

```python
from sklearn.linear_model import ElasticNet
e_model = ElasticNet()
e_model.fit(xtrain,ytrain)
e_accuracy=e_model.score(xtest,ytest)
print(f'The accuracy of the elastic net regression model is {e_accuracy}' )
```

# **Using Decision Tree**

```python
from sklearn.tree import DecisionTreeRegressor
t_model = DecisionTreeRegressor()
t_model.fit(xtrain,ytrain)
t_accuracy=t_model.score(xtest,ytest)
print(f'The accuracy of the random forest regression model is {t_accuracy}' )
```

# **Using KNN**

```python
from sklearn.neighbors import KNeighborsRegressor
k_model = KNeighborsRegressor()
k_model.fit(xtrain,ytrain)
k_accuracy=k_model.score(xtest,ytest)
print(f'The accuracy of the random forest regression model is {k_accuracy}' )
```

# **Using Gradient Boost Regressoor**

```python
from sklearn.ensemble import GradientBoostingRegressor
g_model = GradientBoostingRegressor()
g_model.fit(xtrain,ytrain)
g_accuracy=g_model.score(xtest,ytest)
print(f'The accuracy of the random forest regression model is {g_accuracy}' )
```

```python
print(f'''
****************************************************
                    Summary
****************************************************
LinearRegression Accuracy = {int(lr_accuracy*100)}%
Ridge Accuracy = {int(r_accuracy*100)}%
RandomForest Accuracy = {int(rf_accuracy*100)}%
ElasticNet Accuracy = {int(e_accuracy*100)}%
DecionTree accuracy = {int(t_accuracy*100)}%
KNN accuracy = {int(k_accuracy*100)}%
Gradient Boosting Accuracy = {int(g_accuracy*100)}%
''')
```
