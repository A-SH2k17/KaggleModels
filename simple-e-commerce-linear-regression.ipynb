```python
df = pd.read_csv('/kaggle/input/focusing-on-mobile-app-or-website/Ecommerce Customers')
df
```

```python
from sklearn.model_selection import train_test_split
x = np.array(df['Length of Membership']).reshape(-1,1)
y=df['Yearly Amount Spent']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.9)
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
train_predictions = model.predict(x_train)
test_predictions = model.predict(x_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,train_predictions,color='blue')
plt.xlabel('Length of Membership in Training Set')
plt.ylabel('Yearly Amount Spent in Training Set')
plt.show()
```

```python
plt.scatter(x_test,y_test,color='green')
plt.plot(x_test,test_predictions,color='yellow')
plt.xlabel('Length of Membership in Test Set')
plt.ylabel('Yearly Amount Spent in Test Set')
plt.show()
```

```python
prediction = model.predict([[2]])
print(f'Based on a length of membership of 2 the yearly amount spent is {prediction}')
```
